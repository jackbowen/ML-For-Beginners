```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model, model_selection
```

Now we're going to load in the dataset containing diabetes data. There should be 442 entries, each of which contain 11 columns; 10 of which are predictive factors and the 11th is the progression of the disease after 1 year. The following lines load the first 10 factors into a numpy array called X and the target into one called y.

```python
X, y = datasets.load_diabetes(return_X_y=True)
print(X.shape)
print(X[0])
```

The 10 columns shown above are age in years, sex, body mass index (bmi), average blood pressure (bp), total serum cholesterol (s1 tc), low-density lipoproteins (s2 ldl), high-density lipoproteins (s3 hdl), total cholestoral / HDL (s4 tch), possibly log of serum triglycerides level (s5 ltg), and blood sugar level (s6 glu). Now if you look at the above features for the sample entry that's been printed out, all of the values are unitless numbers with a relatively small absolute value. These columns aren't the raw data. Instead, each of these 10 variables have been centered and scaled by the standard deviation times n_samples (i.e. the sum of the squares of each column totals 1)

The code below is just separating out the age column I think?

```python
X = X[:, np.newaxis, 2]
print(X.shape)
print(X[0])
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.33)
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
plt.scatter(X_test, y_test,  color='black')
plt.plot(X_test, y_pred, color='blue', linewidth=3)
plt.xlabel('Scaled BMIs')
plt.ylabel('Disease Progression')
plt.title('A Graph Plot Showing Diabetes Progression Against BMI')
plt.show()
```
